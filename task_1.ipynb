```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.pipeline import make_pipeline
import seaborn as sns
import matplotlib.pyplot as plt
```

```python


data = pd.read_csv('sms.csv', encoding='latin-1')
data.head()
```

```python
#loading the data from csv file to a pandas dataframe
ds=pd.read_csv('sms.csv', encoding='latin-1')
```

```python
ds
```

```python
#replace the null values with a null string
sms_data=ds.where((pd.notnull(ds)),'')
sms_data.head()
```

```python
#checking the number of rows and columns in the dataframe
sms_data.shape
```

```python
sms_data.info()
```

```python
sms_data.describe()
```

```python
sms_data.loc[sms_data['v1'] == 'spam', 'v1'] = 0
```

```python
sms_data.loc[sms_data['v1']=='ham','v1',]=1
```

```python
#separating the data as texts and label
X=sms_data['v2']
Y=sms_data['v1']
```

```python
X
```

```python
Y
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2 ,random_state=3)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
print(Y.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
#convert text data into (meaningful value) feature vectors that can be used as input to the logistic regression

feature_extraction=TfidfVectorizer(min_df=1,stop_words=['english'],lowercase=True)
X_train_features=feature_extraction.fit_transform(X_train)
X_test_features=feature_extraction.transform(X_test)
```

```python
#convert Y_train and y_test value into integers
Y_train=Y_train.astype('int')
Y_test=Y_test.astype('int')
```

```python
 print(X_train_features)
```

```python
print(X_train_features.shape)
```

```python
print(X_test_features)
```

```python
print(X_test_features.shape)
```

```python
model=LogisticRegression()
```

```python
#training the logistic regression model with the training data
model.fit(X_train_features,Y_train)
```

```python
#prediction on training data
prediction_on_training_data=model.predict(X_train_features)
accuracy_on_training_data=accuracy_score(Y_train,prediction_on_training_data)
```

```python
print('Accuracy on training data:',accuracy_on_training_data)
```

```python
#prediction on testing data
prediction_on_test_data=model.predict(X_test_features)
accuracy_on_test_data=accuracy_score(Y_test,prediction_on_test_data)
```

```python
print('Accuracy on test data : ', accuracy_on_test_data)
```

```python
# Make predictions
predictions = model.predict(X_test_features)

# Evaluate the model
accuracy = accuracy_score(Y_test, predictions)
conf_matrix = confusion_matrix(Y_test, predictions)
classification_rep = classification_report(Y_test, predictions)

print(f"Accuracy: {accuracy}")
print(f"Confusion Matrix:\n{conf_matrix}")
print(f"Classification Report:\n{classification_rep}")
```

```python
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Ham', 'Spam'], yticklabels=['Ham', 'Spam'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
